Q1. Create a program that:

1. Opens a file "names.txt" in write mode

2. Writes 5 names (one per line) entered by the user

3. Then opens the same file in read mode and prints all names

```python
with open("name.txt","w") as file:
    print("EN=nter 5 naames:")
    for i in range(5):
        name = input(f"Name{i+1}: ")
        file.write(name + "\n")

with open("name.txt","r") as file:
    print("\nNames in the file:")
    for line in file:
        print(line.strip())
```

Q3. Create a program that:

1. Has a list of numbers: [5, 10, 15, 20, 25]

2. Uses a list comprehension to create a new list with only numbers greater
than 15

3. Prints the new list

```python
nums=[5,10,15,20,25]
greater_than_15=[num for num in nums if num>15]
print(greater_than_15)
```

Q4. Create a Python dictionary of 3 cities and their populations. Save it to
"cities.json".

1. Then load the JSON and print each city and its population.

2. Ask the user for a new city & its population - update this info in the json
file.

```python
import json
# Create dictionary of cities and populations
cities={
    "Kota": 1000000,
    "Jaipur": 3000000,
    "Udaipur": 500000
}
# Save using json.dumps()
with open("cities.json","w") as file:
    json_string=json.dumps(cities, indent=4)
    file.write(json_string)

#load using json.loads() method
with open("cities.json","r") as file:
    data=json.load(file)

print("Cities and their populations:")
for city, population in data.items():
    print(f"{city}: {population}")

# ask user for city and population
new_city=input("Enter a city name: ")
new_population=int(input("Enter the population: "))

#update dictionary
data[new_city]=new_population

#save updated dictionary back to file
with open("cities.json","w") as file:
    file.write(json.dumps(data, indent=4))

```
