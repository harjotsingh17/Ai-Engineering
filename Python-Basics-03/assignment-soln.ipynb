Q1. Palindrome Checker

```python
def is_palindrome(s):
    s = s.lower().replace(" ", "")  # Remove spaces and convert to lowercase
    reversed_s = s[::-1]
    return s == reversed_s

user_input = input("Enter a string: ")
if is_palindrome(user_input):
    print("It is a palindrome!")
else:
    print("It is not a palindrome!")
```

Q2. Average of List

```python
numbers = list(map(int, input("Enter numbers separated by space: ").split()))
average = sum(numbers) / len(numbers)
print(f"Average: {average}")
```

Q4. Even/Odd Tuples

```python
# You use split() alone when you just need to break a string into a list of words or parts and don't need to change their data type (they stay as strings). 
# You use map() when you already have an iterable

numbers=tuple(map(int, input("Enter number with comma").split(",")))
even_nums = tuple(num for num in numbers if num % 2 == 0)
odd_nums = tuple(num for num in numbers if num % 2 != 0)
print("Even tuple:", even_nums)
print("Odd tuple:", odd_nums)
```

Q5. Student Dictionary Menu

```python
students={}

while True:
    print("\n a-Add, B- Update, C-Search, D-Delete, Q-Quit")
    choice=input("Enter your choice: ").upper()

    if choice=="A":
        name=input("Enter student name:")
        marks=int(input("Enter student marks:"))
        students[name]=marks
        print("Student added successfully!")

    elif choice=="B":
        name=input("Enter student name to update:")
        if name in students:
            marks=int(input("Enter new marks:"))
            students[name]=marks
            print("Student marks updated successfully!")
        else:
            print("Student not found!")
    
    elif choice=="C":
        name=input("Enter student name to search:")
        if name in students:
            print(f"{name} has marks: {students[name]}")
        else:
            print("Student not found!")\
            
    elif choice=="D":
        name=input("Enter student name to delete:")
        if name in students:
            del students[name]
            print("Student deleted successfully!")
        else:
            print("Student not found!")

    elif choice=="Q":
        print("Exiting the program. Goodbye!")
        break
```

Q8. Check Disjoint Lists

```python
list1 = list(map(int, input("Enter first list: ").split()))
list2 = list(map(int, input("Enter second list: ").split()))

set1 = set(list1)
set2 = set(list2)

if set1.isdisjoint(set2):
    print("Lists are disjoint (no common elements)")
else:
    print("Lists share common elements")
```

Q9. Duplicate Elements

```python
numbers = list(map(int, input("Enter list numbers: ").split()))
seen = set()
duplicates = set()

for num in numbers:
    if num in seen:
        duplicates.add(num)
    else:
        seen.add(num)

print("Duplicates:", list(duplicates) if duplicates else "No duplicates")
```

Q10. Unique Characters

```python
text = input("Enter a string: ")
unique_chars = set(text)
print("Unique characters:", ''.join(sorted(unique_chars)))
print("Count of unique characters:", len(unique_chars))
```
